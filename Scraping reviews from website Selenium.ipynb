# Connexion to the url

```python
def connect_to_base(browser,page_number):
    base_url = 'https://www.indeed.fr/cmp/IBM/reviews?fcountry=ALL'+'&start='+str(page_number)
    connection_attempts = 0
    while connection_attempts < 3:
        try:
            browser.get(base_url)
            WebDriverWait(browser, 5).until(EC.presence_of_element_located((By.ID, 'cmp-content')))
            return True
        except Exception as ex:
            connection_attempts += 1
            print("Error connecting to {0}".format(base_url))
            print("Attempt #{0}".format(connection_attempts))
    return False
```

# Parse html link and return database

```python
def parse_html(html):
    
    soup = BeautifulSoup(html,'html.parser')
    blocks = soup.find_all('div',class_='cmp-review-container')
    output_list = []
    
    for link in blocks:
        title = link.find('div',class_='cmp-review-title').get_text()
        rate  = link.span.meta.get("content")     #link.find('div',class_='cmp-ratingNumber').get_text()
        job   = link.find('div',class_='cmp-review-subtitle').get_text()
        desc  = link.find('span',class_='cmp-review-text').get_text()
        ville = link.find('span',class_='cmp-reviewer-job-location').get_text()
        jour  = link.find('span',class_='cmp-review-date-created').get_text()
        page_info = {
            'title': title,
            'rate': rate,
            'job': job,
            'ville': ville,
            'jour': jour,
            'desc': desc
        }
        output_list.append(page_info)
    return(output_list)
```

# Export database

 # First option : 

```python
def write_to_file(output_list, filename):
    for row in output_list:
        with open(filename,'a',newline='',encoding="utf-8") as csvfile:
            fieldnames = ['title', 'rate', 'job', 'ville', 'jour', 'desc']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writerow(row)
```

```python
def run_process(page_number, filename, browser):
    if connect_to_base(browser, page_number):
        sleep(2)
        html = browser.page_source
        output_list = parse_html(html)
        write_to_file(output_list, filename)
    else:
        print('Error connecting to indeed')
```

```python
ILLEGAL_CHARACTERS_RE = re.compile(r'[\000-\010]|[\013-\014]|[\016-\037]|[\x00-\x1f\x7f-\x9f]|[\uffff]')
def illegal_char_remover(data):
    """Remove illegal character"""
    if isinstance(data, str):
        return ILLEGAL_CHARACTERS_RE.sub("", data)
    else:
        return data

target_df = df_final.applymap(illegal_char_remover)
```
